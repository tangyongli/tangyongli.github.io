

# Axis
Matplotlib uses Axes to refer to the drawing area that contains data, x- and y-axis, ticks, labels, title<br>
in the picture, the axis is creare with 'ax = fig.subplots()'. Everything else on the figure was created with methods on this ax object, or can be accessed from it.



### The way to create one axis
An Axes object 封装 all the elements of an individual (sub-)plot in a figure. It contains most of the (sub-)plot elements: Axis, Tick, Line2D, Text, Polygon, etc., and sets the coordinate system.

`ax = fig.subplots()`<br>
`fig.add_axes([0, 0, 1, 1])`<br>
`pyplot.subplot_mosaic` and `Figure.subplot_mosaic`: add a grid of named Axes and return a dictionary of axes.<br>
`fig, axs = plt.subplot_mosaic([['left', 'right'], ['bottom', 'bottom']]), axs['left']`

```python
# axis layout
fig, axs = plt.subplots(ncols=4, figsize=(7, 2.5), layout='constrained')
np.random.seed(19680801)
t = np.arange(200)
x = np.cumsum(np.random.randn(200))
axs[0].plot(t, x)
axs[0].set_title('aspect="auto"')

axs[1].plot(t, x)
axs[1].set_aspect(3) # show different width of subplots
axs[1].set_title('aspect=3')
axs[2].plot(t, x)
axs[2].set_title('aspect="auto"')
axs[3].plot(t, x)
axs[3].set_title('aspect="auto"')
```
